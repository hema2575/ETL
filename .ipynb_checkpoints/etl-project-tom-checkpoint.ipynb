## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
import sys
purchase_data_no_dups_df=purchase_data_df.drop_duplicates(subset=["SN"])
age_demogs_df=pd.DataFrame()
age_series=pd.cut(
    purchase_data_no_dups_df['Age'], 
    bins=[0,9,14, 19, 24, 29, 34,39, sys.maxsize], 
    labels=['<10','10-14','15-19', '20-24', '25-29', '30-34', '35-39', '40+']
)
age_demogs_df["Age Categories"]=age_series
total_count_by_cat=age_demogs_df.groupby(["Age Categories"]).size()
cum_total_count_by_cat=total_count_by_cat.sum()
percentage_of_counts=(total_count_by_cat / cum_total_count_by_cat) * 100
age_demogs_df = pd.DataFrame({"Total Counts":total_count_by_cat,"Percentage of Players":percentage_of_counts})
age_demogs_df["Percentage of Players"]=age_demogs_df["Percentage of Players"].map("{:.2f}%".format)
age_demogs_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchase_df=pd.DataFrame()
age_series=pd.cut(
    purchase_data_df['Age'], 
    bins=[0,9,14, 19, 24, 29, 34,39, sys.maxsize], 
    labels=['<10','10-14','15-19', '20-24', '25-29', '30-34', '35-39', '40+']
)
age_purchase_df["Age Ranges"]=age_series
age_purchase_df["Price"]=purchase_data_df["Price"]
age_purchase_df["Purchase ID"]=purchase_data_df["Purchase ID"]
age_purchase_df["SN"]=purchase_data_df["SN"]
purchase_count1=age_purchase_df.groupby(["Age Ranges"]).size()
average_purchase_price1=age_purchase_df.groupby(["Age Ranges"]).mean()["Price"]
total_purchase_value1=age_purchase_df.groupby(["Age Ranges"]).sum()["Price"]
total_purchasers1=age_purchase_df.drop_duplicates(subset=["SN"]).groupby(["Age Ranges"]).count()["Purchase ID"]
avg_total_purchase_per_person1=total_purchase_value1 / total_purchasers1
purch_by_age_df = pd.DataFrame({"Purchase Count":purchase_count1,"Average Purchase Price":average_purchase_price1,
                                      "Total Purchase Value":total_purchase_value1,"Avg Total Purchase per Person":avg_total_purchase_per_person1})
purch_by_age_df["Average Purchase Price"]=purch_by_age_df["Average Purchase Price"].map("${:.2f}".format)
purch_by_age_df["Total Purchase Value"]=purch_by_age_df["Total Purchase Value"].map("${:,.2f}".format)
purch_by_age_df["Avg Total Purchase per Person"]=purch_by_age_df["Avg Total Purchase per Person"].map("${:.2f}".format)
purch_by_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count=purchase_data_df.groupby(["SN"]).count()["Purchase ID"]
average_purchase_price=purchase_data_df.groupby(["SN"]).mean()["Price"]
total_purchase_value=purchase_data_df.groupby(["SN"]).sum()["Price"]
total_purchasers=purchase_data_df.drop_duplicates(subset=["SN"]).groupby(["SN"]).count()["Purchase ID"]
avg_total_purchase_per_person=total_purchase_value / total_purchasers
top_spenders_df = pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":average_purchase_price,
                                      "Total Purchase Value":total_purchase_value})
top_spenders_df.sort_values(by=['Total Purchase Value'], inplace=True,ascending=False)
top_spenders_df["Average Purchase Price"]=top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df["Total Purchase Value"]=top_spenders_df["Total Purchase Value"].map("${:,.2f}".format)
top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_popular1_df=pd.DataFrame()
most_popular1_df["Item ID"]=purchase_data_df["Item ID"]
most_popular1_df["Item Name"]=purchase_data_df["Item Name"]
most_popular1_df["Price"]=purchase_data_df["Price"]
item_name=most_popular1_df.groupby(["Item ID","Item Name"]).count()
purchase_count2=most_popular1_df.groupby(["Item ID","Item Name"]).size()
item_price=most_popular1_df.drop_duplicates(subset=["Item ID"]).groupby(["Item ID","Item Name"]).min()["Price"]
total_p_value=item_price * purchase_count2
most_popular1_grouped_df = pd.DataFrame({"Purchase Count":purchase_count2,"Item Price":item_price,"Total Purchase Value":total_p_value})
most_popular1_grouped_df.sort_values(by=['Purchase Count'], inplace=True,ascending=False)
most_popular1_grouped_df ["Item Price"]=most_popular1_grouped_df ["Item Price"].map("${:.2f}".format)
most_popular1_grouped_df ["Total Purchase Value"]=most_popular1_grouped_df ["Total Purchase Value"].map("${:.2f}".format)
most_popular1_grouped_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_popular2_df=pd.DataFrame()
most_popular2_df["Item ID"]=purchase_data_df["Item ID"]
most_popular2_df["Item Name"]=purchase_data_df["Item Name"]
most_popular2_df["Price"]=purchase_data_df["Price"]
item_name2=most_popular2_df.groupby(["Item ID","Item Name"]).count()
purchase_count2=most_popular2_df.groupby(["Item ID","Item Name"]).size()
item_price2=most_popular2_df.drop_duplicates(subset=["Item ID"]).groupby(["Item ID","Item Name"]).min()["Price"]
total_p_value2=item_price2 * purchase_count2
most_popular2_grouped_df = pd.DataFrame({"Purchase Count":purchase_count2,"Item Price":item_price2,"Total Purchase Value":total_p_value2})
most_popular2_grouped_df.sort_values(by=['Total Purchase Value'], inplace=True,ascending=False)
most_popular2_grouped_df ["Item Price"]=most_popular2_grouped_df ["Item Price"].map("${:.2f}".format)
most_popular2_grouped_df ["Total Purchase Value"]=most_popular2_grouped_df ["Total Purchase Value"].map("${:.2f}".format)
most_popular2_grouped_df.head()
```
